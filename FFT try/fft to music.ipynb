```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# 读取音频文件
input_file = "audio.wav"
y, sr = librosa.load(input_file)

# 频率转换
target_frequency = 890  # 目标频率
original_frequency = 500  # 原始频率
y_shifted = librosa.effects.pitch_shift(y,sr=sr,n_steps=int(np.round(librosa.hz_to_midi(target_frequency - original_frequency))))

# 计算原始音频的频谱
D_orig = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
# 计算转换后音频的频谱
D_shifted = librosa.amplitude_to_db(np.abs(librosa.stft(y_shifted)), ref=np.max)

# 创建动画
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

def update(frame):
    axs[0].cla()
    axs[1].cla()
    axs[0].set_title('Original Frequency Spectrum')
    axs[1].set_title('Shifted Frequency Spectrum')
    librosa.display.specshow(D_orig, sr=sr, x_axis='time', y_axis='log', ax=axs[0])
    librosa.display.specshow(D_shifted, sr=sr, x_axis='time', y_axis='log', ax=axs[1])

ani = animation.FuncAnimation(fig, update, frames=len(y)//(sr//10), interval=100)

# 保存为视频
ani.save('frequency_comparison.gif', writer='pillow', fps=10)

plt.show()
```
