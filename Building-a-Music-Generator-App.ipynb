## 引入資料

```python
from diffusers import AudioLDM2Pipeline
```

```python
import torch
# import intel_extension_for_pytorch as ipex
model_id = "cvssp/audioldm2"

pipe = AudioLDM2Pipeline.from_pretrained(
    model_id,
    #torch_dtype = torch.float16,
).to("cpu")
```

```python
prompt = "A cheerful ukulele strumming in a beachside jam."
```

```python
generator = torch.Generator("cpu").manual_seed(0)
```

```python
audio = pipe(
    prompt,
    audio_length_in_s = 10.24,
    generator = generator,
).audios[0]
```

```python
from IPython.display import Audio
Audio(audio, rate = 16000)
```

```python
negative_prompt = "Low quality"
```

```python
audio = pipe(
    prompt,
    negative_prompt = negative_prompt,
    audio_length_in_s = 10.24,
    generator = generator,
).audios[0]
```

```python
from IPython.display import Audio
Audio(audio, rate = 16000)
```

## WAV存檔嘗試(OK)

```python
import scipy.io.wavfile as wav

# Assuming `audio` is your numpy array containing audio data
rate = 16000  # Sampling rate
wav.write("musicgen_out.wav", rate, audio)
```

## Gradio 介面

```python
#! pip install -q gradio
```

```python
def create_music(prompt):
  negative_prompt = "Low quality"
  audio = pipe(
    prompt,
    negative_prompt = negative_prompt,
    audio_length_in_s = 10.24,
    generator = generator,
  ).audios[0]
  return 16000, audio
```

```python
import gradio as gr

interface = gr.Interface(
    title = "Music Generation App",
    examples = ["A cheerful ukulele strumming in a beachside jam."],
    fn = create_music,
    inputs = gr.Textbox(),
    outputs = "audio",
).launch(debug=True, share=True)
```
