```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import fft

def plot_frequency_spectrum(wav_file):
    # 讀取.wav檔案
    sample_rate, data = wavfile.read(wav_file)

    # 將數據轉換為單聲道（如果是多聲道）
    if len(data.shape) > 1:
        data = data.mean(axis=1)

    # 計算FFT
    fft_data = fft(data)

    # 計算頻率軸
    frequencies = np.fft.fftfreq(len(data), 1/sample_rate)

    # 繪製頻譜圖
    plt.figure(figsize=(10, 6))
    plt.plot(frequencies[:len(data)//2], np.abs(fft_data[:len(data)//2]))  # 繪製正半軸頻率範圍
    plt.title("Frequency Spectrum")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")
    plt.xlim(0, 3000)  # 如果你只想關注特定範圍的頻率，可以調整x軸的範圍
    plt.grid(True)
    plt.show()

# 使用例子
wav_file = "一閃一閃小星星 (Twinkle Twinkle Little Star) 鋼琴教學 (MEDIUM).wav"  # 替換為你的.wav檔案的路徑
plot_frequency_spectrum(wav_file)
```
